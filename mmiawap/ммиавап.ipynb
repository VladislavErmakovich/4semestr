```python
import math
```

```python
def resonance_frequency(a, S, h, E, g, p, m):

  f = (math.pi/2) * ((4 * (a**4) + S**2)/(4 * (a**2) * (S**2))) * math.sqrt((E * (h**3))/(12 * (1 - g**2)*(p*h + m/S)))

  return f
```

```python
def df_da(a):
  return (math.pi/2) * (((4 * (a**4)) - S**2)/(2 * (a**3) * (S**2))) * math.sqrt((E * h**3)/(12 * (1 - g**2)*(p*h + m/S)))
```

```python
E = 3.3 * 10**10
g = 0.279
p = 2.47 * 10**3
m = 50/1000
S = 0.03
max_a = 0.3
h = 0.003
```

```python
a = 0.174
iter = 1
k = 100
while True:
  f = resonance_frequency(a, S, h, E, g, p, m)
  da = df_da(a)
  print(f'{iter}\t{a}\t{S/a}\t{h}\t{da}\t{f}\t{k}\n')

  new_a = a + k*da
  new_f = resonance_frequency(new_a, S, h, E, g, p, m)
  if k<=0.00000001:
    break
  if new_a <= max_a and new_f > f:
    f = new_f
    a = new_a
  else:
    k /= 2
  iter += 1



```

```python
print(f'Лучшее значения\nДлина: {a} м\nШирина: {S/a} м\nВысота: {h} м\nЗначение частоты: {f} Гц\nКоличество итераций: {iter}')
```
